```python
# Scraping data
merk = []
product_name = []
prices = []
ratings = []
urls = []
for page in range(1,11):    # untuk page dari range 1-10 // halaman 1-10
    driver.get(f'https://www.zalora.co.id/search?q=tas&page={page}')         # mengambil data dari web Zalora
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # memberikan perintah untuk scroll halaman langsung sampai ke bawah
    time.sleep(5) # memberi jeda 10 detik sebelum pindah halaman
    html = driver.page_source 
    soup = BeautifulSoup(html, 'html.parser')
    for data in soup.find_all('div', {'class':'flex flex-col min-h-[358px] tablet:min-h-[380px] desktop:min-h-[500px]'}): # mengambil semua data di soup
        try:
            merk.append(data.find("span", {"class":"shrink-[10] truncate text-base font-bold"}).get_text()) # mengambil semua data merk
        except:
            merk.append(None)           #diisi none jika tidak ada data
        try:
            product_name.append(data.find("div", {"class":"overflow-hidden text-ellipsis whitespace-nowrap text-sm"}).get_text()) # mengambil semua data nama produk
        except:
            product_name.append(None)   #diisi none jika tidak ada data
        try:
            prices.append(data.find("div", {"class":"text-grey-60 line-through"}).get_text()) # mengambil semua data harga
        except:
            prices.append(None)         # diisi none jika tidak ada data
        try:
            ratings.append(data.find("span", {"class":"text-xxs font-medium leading-3 text-grey-60"}).get_text()) # mengambil semua data rating
        except:
            ratings.append(None)        #diisi none jika tidak ada data
        for link in soup.find_all("a", {"class": "relative"}):
            try:
                # Ambil atribut href dan gabungkan dengan URL base
                href = link["href"]
                urls.append(href)
            except KeyError:
                # Jika atribut 'href' tidak ditemukan, tambahkan None
                urls.append(None) 
```

```python
# Cek Length Column
len(merk),len(product_name), len(prices), len(ratings), len(urls)
```

```python
# Menyamakan panjang list

# Mencari panjang minimum
min_length = min(len(merk), len(product_name), len(prices), len(ratings), len(urls))

# Menyesuaikan semua list ke panjang minimum
merk = merk[:min_length]
product_name = product_name[:min_length]
prices = prices[:min_length]
ratings = ratings[:min_length]
urls = urls[:min_length]
```

```python
# Cek Length Column setelah menyesuaikan panjang list
len(merk),len(product_name), len(prices), len(ratings), len(urls)
```

```python
# Create dataframe
df = pd.DataFrame({
                'Merk': merk,
                'Nama_Produk': product_name,
                'Harga': prices,
                'Ratings': ratings,
                'URL': urls
})
df.head(10)
```

```python
# Menghapus "Rp" dan "." pada kolom "Harga"
df['Harga'] = df['Harga'].str.replace('\D', '', regex=True)


# \D: Ini adalah ekspresi reguler (regex) yang berarti "semua karakter yang bukan angka" (non-digit).
# regex=True: Ini menunjukkan bahwa kita menggunakan ekspresi reguler (regex) dalam pencarian dan penggantian karakter. 
```

```python
# Cek data setelah dihapus
df.head(10)
```

```python
# Cek informasi data
df.info()
```

```python
# Mengganti Tipe Data (menyesuaikan)

# Mengganti tipe data kolom harga (object to int64)
df['Harga'] = df['Harga'].fillna('0')      # Ganti NaN dengan '0' jika ada
df['Harga'] = df['Harga'].astype(int)

# Mengganti tipe data kolom ratings (object to float64)
df['Ratings'] = df['Ratings'].fillna('0')  # Ganti NaN dengan '0' jika ada
df['Ratings'] = df['Ratings'].astype(float)
```
